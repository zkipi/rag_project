```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=["Title", "Description", "Type"])
```

```python
df["Director"] = df["Director"].fillna("Unknown")
df["Cast"] = df["Cast"].fillna("Unknown")
df["Country"] = df["Country"].fillna("Unknown")
df["Rating"] = df["Rating"].fillna("Unknown")
df["Duration"] = df["Duration"].fillna("Unknown")
```

```python
df["combined_text"] = (
    "Title: " + df["Title"].astype(str) +
    ". Category: " + df["Category"].astype(str) +
    ". Genre: " + df["Type"].astype(str) +
    ". Director: " + df["Director"].astype(str) +
    ". Cast: " + df["Cast"].astype(str) +
    ". Country: " + df["Country"].astype(str) +
    ". Release date: " + df["Release_Date"].astype(str) +
    ". Rating: " + df["Rating"].astype(str) +
    ". Duration: " + df["Duration"].astype(str) +
    ". Description: " + df["Description"].astype(str)
)
```

```python
df["Release_Date"] = df["Release_Date"].fillna("Unknown")
```

```python
df.to_csv("../data/processed/cleaned_netflix.csv", index=False)
```

# Sammanfattning

Datasetet har undersökts och städats innan det används i vårt RAG-system.

Jag:
- kontrollerade datatyper
- hanterade saknade värden
- kontrollerade dubbletter
- skapade en combined_text-kolumn för embeddings och retrieval

Detta förbättrar datakvaliteten och gör det lättare för RAG-systemet att hitta relevant information.
